1. Import required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Load the dataset

```python
df = sns.load_dataset("titanic")
```

3. Scan all numeric variables in dataset

```python
numeric_variables = df.select_dtypes(include=['int','float']).columns.tolist()
```

```python
numeric_variables
```

4. Scan numeric variables for outliers using box plots

```python
for variable in numeric_variables:
  # Create a box plot for variable 
  sns.boxplot(x=df[variable])
  plt.title(f'Box plot for : {variable}')
  plt.show()
  print("\n")

  # Identify outliers based on the box plot
  q1 = df[variable].quantile(0.25)
  q3 = df[variable].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr

  # Handle outliers using different techniques
  # a) Min Max Normalization 
  min_value = df[variable].min()
  max_value = df[variable].max()
  df[variable+'_minmax'] = (df[variable] - min_value) / (max_value - min_value)

  # b) Z score Normalization 
  mean = df[variable].mean()
  std_dev = df[variable].std()
  df[variable+'_zscore'] = (df[variable] - mean) / std_dev

  # c) Remove Outliers
  df[variable+'_no_outliers'] = df[variable][((df[variable] > lower_bound) & (df[variable] < upper_bound))]
```

# Assignment 7 (Data Analytics I)

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
```

```python
# Split the dataset into features (X) and target variable (Y)
print(df.columns)
X = df['lstat'].values.reshape(-1, 1) # input features (Linear Regression) 
# X = df.drop('medv',axis=1) # input features (Multiple Regression)
Y = df['medv'] # target variable 
```

```python
# Split the data into training and test sets (25% Testing , 75% Training)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=5)
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Fit the model to the training data
model.fit(X_train,Y_train)
```

```python
# Make predictions on the test data
Y_pred = model.predict(X_test)
```

```python
# Evaluate the model 
rmse = np.sqrt(mean_squared_error(Y_test,Y_pred)) # or rmse = mean_squared_error(Y_test,Y_pred,squared=False)
r2 = r2_score(Y_test,Y_pred)*100
```

```python
print("Root Mean Squared Error : ",rmse)
print("r2 Score (%): ",r2)
```

# Assignment 8 (Data Analytics II)

```python
# Importing required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

import warnings
warnings.filterwarnings("ignore") # used to ignore warnings 
```

```python
# Importing dataset
df = pd.read_csv("https://raw.githubusercontent.com/shivang98/Social-Network-ads-Boost/master/Social_Network_Ads.csv")
```

```python
df
```

```python
# Separate the features and the target variable (Split the dataset)
X = df[['Age','EstimatedSalary']]
Y = df['Purchased']
```

```python
# Splitting the dataset into training and testing sets (75% training, 25% testing)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
# Create an instance of LogisticRegression
model = LogisticRegression()
```

```python
# Fit the model on the data
model.fit(X_train,Y_train)
```

```python
# Perform predictions on the data
Y_pred = model.predict(X_test)
```

```python
# Compute the confusion matrix
conf_matrix = confusion_matrix(Y_test,Y_pred)
```

```python
TN = conf_matrix[0][0]
FP = conf_matrix[0][1]
FN = conf_matrix[1][0]
TP = conf_matrix[1][1]
```

```python
# Compute accuracy, error rate, precision, and recall 

accuracy = accuracy_score(Y_test, Y_pred) #(TP + TN) / (TP+TN+FP+FN)
print("Accuracy : ",accuracy)
```

```python
precision = precision_score(Y_test,Y_pred,average="micro") #(TP) / (TP + FP)
print("Precision : ",precision)
```

```python
recall = recall_score(Y_test,Y_pred,average="micro") # TP / (TP+FN)
print("Recall : ",recall)
```

```python
error_rate = 1 - accuracy
print("Error rate : ", error_rate)
```

# Assignment 10 (Text Analysis)

```python
import numpy as np

import nltk
# nltk.download(['punkt','averaged_perceptron_tagger','stopwords','wordnet'])
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
# Sample Document 
document = "The quick brown fox jumps over the lazy dog."
document_2 = "My dog is not lazy."
```

```python
# Tokenization
tokens = word_tokenize(document)
print("Tokens : ",tokens)
```

```python
# POS(Part-of-speech) Tagging
pos_tags = pos_tag(tokens)
print("POS Tags : ",pos_tags)
```

```python
# Stop Words Removal
stop_words = set(stopwords.words("english"))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
print("Filtered Tokens : ",filtered_tokens)
```

```python
# Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
print("Stemmed Tokens : ",stemmed_tokens)
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
print("Lemmatized Tokens : ",lemmatized_tokens)
```

```python
# TF-IDF Representation

# Term Frequency 
tokens = word_tokenize(document)
doc_dict = dict()
for token in tokens:
  if token in doc_dict.keys():
    doc_dict[token]+=1
  else:
    doc_dict[token] = 1
for key,value in doc_dict.items():
  doc_dict[key] = value/len(tokens)
print("TF of first document : ",doc_dict) # Term frequency of all words in document

tokens = word_tokenize(document_2)
doc_dict_2 = dict()
for token in tokens:
  if token in doc_dict_2.keys():
    doc_dict_2[token]+=1
  else:
    doc_dict_2[token] = 1
for key,value in doc_dict_2.items():
  doc_dict_2[key] = value/len(tokens)
print("TF of second document : ",doc_dict_2) # Term frequency of all words in document 2

# Inverse Document Frequency 
total_no_of_docs = 2;
final_dict = dict()
for key in doc_dict.keys():
  if key in final_dict.keys():
    final_dict[key] += 1
  else:
    final_dict[key] = 1;
for key in doc_dict_2.keys():
  if key in final_dict.keys():
    final_dict[key] += 1
  else:
    final_dict[key] = 1;
for key,value in final_dict.items():
  final_dict[key] = np.log(total_no_of_docs/value)
print("IDF : ",final_dict)
```
